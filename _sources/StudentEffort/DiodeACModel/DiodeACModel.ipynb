# Introduction

Diodes are critical components in modern electronic circuits, renowned for their non-linear I-V characteristics. While these properties enable functionalities like rectification and switching, analyzing a diodes behavior for small variations around a fixed operating point requires a different perspective. This brings us to the **small-signal model**a powerful tool that simplifies the diode's non-linear dynamics into an intuitive, linear framework.

This presentation delves into the process of deriving the small-signal model, grounded in the principles of DC biasing and linearization. Using visualizations and equations inspired by Sedra and Smiths analysis, well explore how the bias point (Q-point) allows us to approximate the diodes behavior as a linear system. The model is vital in applications like amplifiers and other analog circuits, where small deviations in input signals must be analyzed with precision.

By comprehending this methodology, we gain a profound understanding of the diode's significance in circuit design and acquire the analytical tools necessary to accurately predict its behavior under small-signal conditions. This presentation aims to bridge the conceptual gap between the diode's inherently non-linear characteristics and its linearized applications in practical circuit analysis.


# Diode Small Signal Model

The figure below (Figure 1.13 from Sedra and Smith 7 ed.) shows the transfer characteristics of a linear amplifier with power supplies $\mathrm{L}_{+}$and $\mathrm{L}_{\text {- }}$. It also shows two sinusoidal inputs to the amplifier plotted vertically, and the corresponding outputs plotted horizontally.


This figure makes that point that if the input signal amplitude is too large, the output will exceed the power supplies. In that case, we are not operating in the linear region, and the output "clips". This is happening for the input labeled 2 , so the output is a function that is more complex than a single sinusoid. That means it has many Fourier components, and from our definition of linearity, the circuit is not linear under those circumstances.

Figure 1.14 below shows the transfer characteristics of an amplifier that is more typical, and more complex. There is a region of the transfer characteristics where the amplifier response is approximately linear, but it is offset from 0 Volts. Therefore, to amplify a signal $v_{i}$ with this amplifier, we must add a DC bias voltage $\mathrm{V}_{1}$ to the signal, to move the signal into a linear region of the transfer characteristics.


Even then, the transfer characteristics may not be precisely linear. If they are not, our output signal will be a distorted version of the input. In real amplifiers, the transfer characteristic is very nearly linear, provided we do not move too far away from the region labeled $Q$ in the figure above. In other words, if the input signal is small in amplitude, then the transfer characteristics will be approximately linear over the range of the signal amplitude, and the output will be a good reproduction of the input.

To summarize: in real amplifiers, we must apply a bias voltage to our signal to get the signal into a region of the transfer characteristics that is approximately linear. Furthermore, the signal must be small, that is, of a small amplitude, so that it does not vary from the linear region of the transfer characteristics.

The figure below shows the transfer characteristics $i_{D}-v_{D}$ of a typical diode. We don't usually think of a diode as an amplifier, but the notion of bias and application of a small signal still applies. In the figure, the bias voltage is $\mathrm{V}_{\mathrm{D}}$, which is generated by the biasing source $\mathrm{V}_{\mathrm{DD}}$. The small signal input is $v_{d}$. The ac component of the output is $i_{d}$, and the DC component is $\mathrm{I}_{\mathrm{D}}$. The quiescent point or bias point $Q$ is the point on the current-voltage curve that is not varying, that is, it's the DC voltage and current $\mathrm{V}_{\mathrm{D}}$ and $\mathrm{I}_{\mathrm{D}}$.


$$
\begin{aligned}
& i_{D}(t)=I_{D}+i_{d}(t) \\
& v_{D}(t)=V_{D}+v_{d}(t)
\end{aligned}
$$


We will use this figure to develop the small signal model for the diode, that is, to develop a circuit model that is appropriate for a small AC signal applied to the linear region of the transfer characteristics.

# Finding the Small Signal Model

Figure 4.14 shows a tangent drawn at the bias point $Q$. This line approximates the transfer characteristics in the case that the signal amplitude is small. The slope of that line is an inverse resistance, and the resistance is $r_{d}$, which we will call the small signal resistance of the diode. The resistance is the inverse slope of the tangent to the curve evaluated at the bias point:
$$
r_{d}=\left(\frac{\partial \mathrm{i}_{D}}{\partial v_{D}}\right)_{v_{D}=V_{D}}^{-1}
$$

If we extrapolate the line to $i_{D}=0$, it intercepts the axis at a value we are calling $\mathrm{V}_{\mathrm{DO}}$. Note that and $r_{d}$ are the parameters of a model for the forward bias diode of the type that we discussed in diode modeling - a voltage source in series with a resistance. But depending upon
the $D C$ bias $V_{D}$ and therefore on the $Q$ point, $r_{d}$ and $V_{D O}$ will change. For example, if $V_{D}$ increases, the slope of the line will increase, or $r_{d}$ will decrease, and $V_{D O}$ will increase.

We develop the small signal model of the diode beginning with the ideal diode equation:
$$
i_{D}=I_{S}\left(e^{v_{D} / n V_{t}}-1\right)
$$

Recall that $\mathrm{I}_{s}$ is the reverse bias saturation current, n is the ideality factor, and $\mathrm{V}_{\mathrm{t}}$ is the thermal voltage, which at room temperature is 25 mV . We will consider forward bias where $v_{D} \gg \mathrm{nV}_{\mathrm{t}}$, in which case we can approximate to
$$
i_{D}=I_{S} e^{v_{D} / n v_{t}}
$$

Now $v_{D}$ has an AC and a DC component: $v_{D}=V_{D}+v_{d}$. Putting that in our equation above gives
$$
i_{D}=I_{S} e^{V_{D} / n V_{t}} e^{V_{d} / n V_{t}}
$$

The first term on the left is the DC current generated by application of $\mathrm{V}_{\mathrm{D}}$ :
$$
I_{D}=I_{S} e^{V_{D} / n V_{t}}
$$

If the signal is small, specifically $v_{d}<<n V_{t}$, we can expand the exponent $e^{V_{d} / n v_{t}}$ using Taylor series expansion:

$$
e^x \simeq 1 + x + \frac{x^2}{2!} + \frac{x^3}{3!} + \cdots \quad
$$
$$
\text{for } (x \ll 1) \quad \Rightarrow \quad e^x \simeq 1 + x
$$

So:

$$
i_{D} \approx I_{D}\left(1+\frac{V_{d}}{n V_{t}}\right)=I_{D}+V_{d} \frac{I_{D}}{n V_{t}}
$$

Then $\mathrm{I}_{\mathrm{D}}$ is the DC component of the current, and the AC component is $\frac{V_{d}}{r_{d}}$, where
$$
r_{d}=\frac{n V_{t}}{I_{D}}
$$

Now we have our diode small signal model. It is a voltage source $\mathrm{V}_{\mathrm{DO}}$ in series with a resistance $r_{d}$, where $r_{d}$ is determined by the biasing conditions, specifically $I_{D}$. So we don't have to evaluate
the derivative of the curve; we can simply use the formula above. But we do have to know the bias point.

Let's look at a simple circuit, and split it into its DC and AC components. To do this, we first replace the diode with the small signal model. This is done for the circuit below. (There are lots of symbols here...make sure you understand what they mean!)


A KVL for the circuit on the right gives
$$
V_{D D}+v_{d}=\left(I_{D}+i_{d}\right)\left(R+r_{d}\right)+V_{D 0}
$$

We can separate the DC components and the AC components:
$$
\begin{aligned}
& \mathrm{DC}: V_{D D}=I_{D}\left(R+r_{d}\right)+V_{D 0}=I_{D} R+V_{D} \\
& \mathrm{AC}: v_{d}=i_{d}\left(R+r_{d}\right)
\end{aligned}
$$

The second of the DC equations comes from recognizing that $I_{D} r_{d}+V_{D 0}=V_{D}$. The DC and AC circuits are shown below.


We have found that the $\boldsymbol{A C}$ model for a diode is a resistor whose value depends on the diode biasing value $\mathrm{I}_{\mathrm{D}}$.

# Example: Find the AC model of the circuit below.



$I_B$ is a dc bias current: $\quad I_B=1.4 \mathrm{~A}$.

$i_s$ is a small signal input = 0.05.

$D$ is a diode:
$$
n V_T=0.15
$$
$$
i_D = 0.05\left(e^{v_D / 0.15}-1\right) \mathrm{A}
$$

We want to find the small signal gain $i_{o} /{ i}_s$.
First we find the dc bias point for D:
$$
\begin{aligned}
i_D=I_D= & 0.05\left(e^{V_D / 0.15}-1\right) = 1.4 \\
& e^{V_D / 0.15}=\frac{1.4}{0.05}+1=29 \\
& V_D=0.15 \operatorname{ln}(29)=0.50516
\end{aligned}
$$

So the dc bias point is $I_D, V_D=1.4 \mathrm{~A}, 0.5051 \mathrm{~V}$.



Now we need the $a c$ resistance rd:
$$
\begin{aligned}
& 1 / r_d=\left.\frac{\partial i_D}{\partial v_D}\right|_{v_D}=\frac{1}{0.15}(0.05) e^\frac{0.5051} {0.15}
=9.667 \\
& \Longrightarrow r_d = 0.103 \Omega
\end{aligned}
$$

We could also have used:
$$
r_d=\frac{n v_T}{i_D}=\frac{0.15}{104}=0.107 \Omega
$$

Then the ac circuit is:



\begin{aligned}
& i_o=-250 v_d \cdot \frac{1}{1+3} \xrightarrow{v_d=i_s r_d}\frac{i_o}{i_s}=-250 \cdot r_d \cdot \frac{1}{4} \\
& \xrightarrow{r_d=0.103} \frac{i_o}{i_s}=-250 \cdot 0.103 \cdot \frac{1}{4}=-6.69
\end{aligned}

so the small signal gain $i_{o} /{ i}_s$ of this circuit is -6.69.

### Steps to Find the AC Model of a Diode

Finding the **AC model of a diode** involves linearizing the diode's behavior around a specific operating point to analyze small-signal behavior. This method is commonly used in small-signal analysis of circuits like amplifiers. Here are the detailed steps:

---

**1. Identify the DC Operating Point**

Before analyzing the AC model, determine the diode's operating point using the DC circuit conditions:

- Voltage $ ( V_D ) $ : The DC voltage across the diode.  
- Current $ ( I_D ) $ : The DC current through the diode.

The diode's I-V characteristic follows the Shockley equation:

$$
I_D = I_S \left( e^{\frac{V_D}{nV_T}} - 1 \right)
$$

where:
- $ I_S $: Saturation current.
- $ V_T $: Thermal voltage $ ( V_T \approx 26 \, \text{mV} ) $ at room temperature).
- $ n $: Ideality factor (typically 12).

---

**2. Linearize the I-V Curve Around the Operating Point**

The AC model is derived by approximating the diode's nonlinear behavior with a straight line near the operating point. Use a **Taylor expansion** for small deviations $ v_d $ around $ V_D $:

$$
i_d \approx I_D + \frac{\partial I_D}{\partial V_D} \cdot v_d
$$

The term $ \frac{\partial I_D}{\partial V_D} $ is the slope of the I-V curve, also known as the **small-signal conductance**.

---

**3. Calculate the Small-Signal Resistance**

The **small-signal resistance** $ r_d $ is the reciprocal of the small-signal conductance:

$$
r_d = \frac{1}{\frac{\partial I_D}{\partial V_D}}
$$

From the Shockley equation:

$$
\frac{\partial I_D}{\partial V_D} = \frac{I_D}{n V_T}
$$

Thus:

$$
r_d = \frac{n V_T}{I_D}
$$

---

**4. Replace the Diode with its AC Equivalent**

In the small-signal model:
- Replace the diode with a **resistor** $ r_d $, which represents the diode's dynamic resistance at the operating point.
- For AC analysis, the DC voltage source is shorted, and the AC current/voltage deviations are analyzed.

---

**5. Incorporate AC Signal Conditions**

Combine $ r_d $ with the surrounding circuit components. Analyze the circuit using standard small-signal analysis techniques like:
- Nodal analysis.
- Mesh analysis.
- Thevenin/Norton equivalent circuits.

---

**6. Validate the Model**

- Ensure the small-signal approximation is valid by keeping the AC signal $ v_d $ small (e.g., $ v_d \ll V_T $).  
- Check if the circuit behavior matches the expected linear behavior under small-signal conditions.
