## Definition of Complex Number
A complex number is any number that can be written in the form $z = x + yi$ where $x$ is called the __real__ part and $yi$ the __imaginary__ part. The complex number $z = x + yi$ is said to be written in real-imaginary form.The imaginary unit i is defined to be i=√−1. The imaginary unit is sometimes labelled j, usually in engineering.

In phasors,imaginary unit labelled by j.

A complex number can be represented in three ways: 

-   Rectangular Form
-   Polar Form
-   Exponential Form

$z = x + yi$ is an example from __Rectangular Form__ . 


examples: 
$$
z = j+1 \\
z = \sqrt 2 + \frac{j}{2}\\\
z = 3+\pi j 
$$

### Some properties of complex numbers

$$
z = x + yj \\
\bar z = x -yj \\
\bar{z} . z = x^2+y^2
$$

### Addition and Subtraction

Given two complex numbers :

$$
z_1 = x_1 + y_1j \\
z_2 = x_2 + y_2j \\
$$

their sum and difference is found by adding or subtracting their real and imaginary parts separately:

$$
(z_1 + z_2) = (x_1 + x_2) + (y_1 + y_2)j \\
(z_1 - z_2) = (x_1 - x_2) + (y_1 - y_2)j \\
$$


### Multiplication

$$
z_1.z_2 = (x_1x_2 - y_1y_2) + (x_1y_2 +y_1x_2)j
$$

example : 

$$
z_1.z_2 = (3-j)(-5+2i) \\
= -13 +11j \\
j.z = j(3-2j) \\
= 2+3j
$$

## Polar Form of Complex Numbers

In polar form, complex numbers are represented as the combination of the modulus r and argument $\theta$ of the complex number.

We write complex numbers in terms of the distance from the origin and a direction (or angle) from the positive horizontal axis. Polar coordinates are expressed as $(r, \theta)$.

The values of polar coordinates and rectangular coordinates depend on each other. If we know any two values, the remaining two values can be found easily using the relation established between them.

Polar form for a complex number $z=x+yj$ :

$$
z = x + jy <=> z = r ∠ \theta \\
r = \sqrt{x^2 + y^2} \\
\theta = \tan^{-1} (\frac{y}{x})
$$

The components of polar form of a complex number are:

-    $r$ : It signifies absolute value or represents the __modulus__ of the complex number.

-    Angle $\theta$ : It is called the __argument__ of the complex number.

example:

-   Find polar form of $z = 5j$
$$
r = \sqrt{0^2 + 5^2} = 5 , \theta = \tan^{-1}\frac{5}{0} = \tan^{-1} ∞ = \frac{\pi}{2}  \\
=> z = 5 ∠ \frac{\pi}{2}
$$

-   diagram of $z = 2-3j$

$$
r = \sqrt{2^2 + (-3)^2} = \sqrt 13 , \theta = \tan^{-1}\frac{-3}{2} = -56.3^\circ
$$

## Exponential Form of Complex Numbers

We have considered complex numbers in the Rectangular Form, ( a + jb ) and the Polar Form, ( A ∠±θ ). Third method for representing a complex number which is similar to the polar form that corresponds to the length (magnitude) and phase angle of the sinusoid but uses the base of the natural logarithm, e = 2.718 281.. to find the value of the complex number. This third method is called the Exponential Form.

The Exponential Form uses the trigonometric functions of both the sine ( sin ) and the cosine ( cos ) values of a right angled triangle to define the complex exponential as a rotating point in the complex plane.

$$
z = re^{j\theta}
$$

### Euler’s formula

Euler’s formula gives us a way of converting a complex number from its exponential form into its rectangular form.

$$
z = re^{j\theta} <=> z = r(\cos{\theta} + j\sin{\theta})
$$

example : Find rectangular form of z
-   $z = 2e^{\frac{\pi}{3}j} $ 
$$
z = 2(\cos\frac{\pi}{3} + j\sin\frac{\pi}{3}) = 1 + j\sqrt 3 
$$

## About Phasor Analys
### Intrudoction

A phasor is a complex number in polar form that you can apply to circuit analysis. When you plot the amplitude and phase shift of a sinusoid in a complex plane, you form a phase vector, or phasor.

We can also think of a phasor as a rotating vector. Unlike a vector having magnitude and direction, a phasor has magnitude 
$V_a$
and angular displacement 
$ϕ$
. You measure angular displacement in the counterclockwise direction from the positive x-axis.

Here is a diagram of a voltage phasor as a rotating vector at some frequency, with its tail at the origin. If you need to add or subtract phasors, you can convert the vector into its x-component (
$V_a\cosϕ$
) and its y-component (
$V_a\sinϕ$
) with some trigonometry.




![RC1_charging](./image/PhasorDiagram.png)


The following sections explain how to find the different forms of phasors and introduce you to the properties of phasors.



### Finding phasor forms

Phasors, which you describe with complex numbers, embody the amplitude and phase of a sinusoidal voltage or current. The phase is the angular shift of the sinusoid, which corresponds to a time shift 
$t0$
. So if you have 
$\cos[ω(t – t_0)]$, then 
$ωt_0 = ϕ_O$, where 
$ϕ_O$
 is the angular phase shift.

$$
\cos[ωt + ϕ]
$$

To establish a connection between complex numbers and sine and cosine waves, you need the complex exponential 
$e^jθ$
 and Euler’s formula:

$$
e^jθ = \cosθ + j\sinθ \\
where:j = \sqrt-1
$$

The left side of Euler’s formula is the polar phasor form, and the right side is the rectangular phasor form. You can write the cosine and sine as follows:

$$
\cosθ = Re[e^jθ] \\
\sinθ = Img[e^jθ]
$$

$Re[e^jθ]$
 is real part of 
$e^jθ$
 and 
$Img[e^jθ]$
 is imagination part of 
$e^jθ$ . 


As a result :

$$
V(t) = V_A\cos(ωt + ϕ)  <=> V = V_A ∠ ϕ
$$

Now we will solve some examples.

-   example : if currunt $I = (20 + j20\sqrt 3)$ A , find $I(t)$


$$
r_I = \sqrt{20^2 + (20\sqrt 3)^2} = 40A \\
θ_I = \tan^{-1}\frac{20\sqrt 3}{20} = \tan^{-1}\sqrt 3 = 60^\circ \\
=> I = 40 ∠ 60^\circ => I(t) = 40\cos(ωt + 60^\circ)
$$


```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

omega = 2 * np.pi * 50  
phi = np.radians(60)     # convert degree to radiant
t = np.linspace(0, 0.1, 1000)  

I_t = 40 * np.cos(omega * t + phi)

plt.figure(figsize=(10, 5))
plt.plot(t, I_t, label='I(t) = 40cos(ωt + 60°)')
plt.title('I(t)')
plt.xlabel('time(second)')
plt.ylabel('currunt(A)')
plt.grid()
plt.axhline(0, color='black', lw=0.5, ls='--')
plt.legend()
plt.show()
```

### Complex Number in RC circuit


This example is a RC circuit

-   example: Find the current in terms of t in the circuit below



```python
import schemdraw
import schemdraw.elements as elm

with schemdraw.Drawing() as d:
    elm.SourceSin().label('cos(t)').up()
    elm.Resistor().label('2Ω').right()
    elm.Capacitor().label('1F').down()
    elm.Line().left()
```

First find $\omega$ and $V_A$  
$$
V(t) = \cos t = 1 ∠ 0^\circ \\
=> V_A = 1 , \phi = 0^\circ , \omega = 1 \\
$$


Now, we draw the shape of the circuit in the __phasor__

```python
import schemdraw
import schemdraw.elements as elm

with schemdraw.Drawing() as d:
    elm.SourceSin().label('1 ∠ 0°').up()
    elm.Resistor().label('2Ω').right()
    elm.Capacitor().label('1/jw').down()
    elm.Line().left()
```

$$
I = \frac{V_A}{R+\frac{1}{j\omega}} \\
=> I = \frac{1}{2+\frac{1}{j\omega}} = \frac{j\omega}{2j\omega+1} \\
=> I = \frac{j}{2j+1} * \frac{1-2j}{1-2j} = \frac{j-2j^2}{1-4j^2} \\
$$
-   $j=\sqrt{-1} => j^2 = -1$
$$
=> I = \frac{j+2}{5}
$$
-   $a + jb = \sqrt{a^2 + b^2} ∠ \tan^{-1} \frac{b}{a} $
$$
=> 2 + j = \sqrt{2^2 + 1^2} ∠ \tan^{-1} \frac{1}{2} = \sqrt 5 ∠ 26.5^\circ \\
=>I = \frac {\sqrt 5 ∠ 26.5^\circ}{5} => I(t) = \frac{\sqrt 5}{5}\cos{(t + 26.5^\circ)}
$$

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import math

omega = 2 * np.pi * 50  
phi = np.radians(26.5)  
t = np.linspace(0, 0.1, 1000) 

I_t = math.sqrt(5)/5 * np.cos(omega * t + phi)

plt.figure(figsize=(10, 5))
plt.plot(t, I_t, label='I(t) = √5/5cos(t + 26.5°)')
plt.title('I(t)')
plt.xlabel('time(second)')
plt.ylabel('currunt(A)')
plt.grid()
plt.axhline(0, color='black', lw=0.5, ls='--')
plt.legend()
plt.show()
```

### Complex Number in RL circuit


This example is a RL circuit

-   example: Find the current in terms of t in the circuit below($V_s = 3\sin({2t+\frac{\pi}{2}})$)

```python
import schemdraw
import schemdraw.elements as elm

with schemdraw.Drawing() as d:
    elm.SourceSin().label('Vs').up()
    elm.Resistor().label('2Ω').right()
    elm.Inductor().label('3H').down()
    elm.Line().left()
```

First find $\omega$ and $V_A$

-   $z = re^jθ = r(\cos θ + j\sin θ)$
$$
V_s = 3\sin({2t+\frac{\pi}{2}}) = 3e^{j\frac{\pi}{2}} \\
=> V_A = 3 , \phi = \frac{\pi}{2} , \omega = 2 \\
$$


Now, we draw the shape of the circuit in the __phasor__

```python
import schemdraw
import schemdraw.elements as elm

with schemdraw.Drawing() as d:
    elm.SourceSin().label('3 ∠ 0°').up()
    elm.Resistor().label('2Ω').right()
    elm.Inductor().label('3jw').down()
    elm.Line().left()
```

$$
I = \frac{V_A}{R + Lj\omega} \\
=> I = \frac{3}{2 + 3j\omega} = \frac{3}{2 + 6j} \\
=> I = \frac{3}{2 + 6j} * \frac{2 - 6j}{2 - 6j} \\
=> I = \frac{6-18j}{4 - 36j^2} = \frac{3-9j}{20} \\
=> 3 - 9j = \sqrt{3^2 + (-9)^2} ∠ \tan^{-1}{ \frac{-9}{3}} = \sqrt{90} ∠ -71.5^\circ\\ 
=> I = \frac{\sqrt{90} ∠ -71.5^\circ}{20} => I(t) = \frac{\sqrt{90}}{20}\sin{(2t - 71.5^\circ)}
$$

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import math

omega = 4 * np.pi * 50  
phi = np.radians(-71.5)    
t = np.linspace(0, 0.1, 1000) 

I_t = math.sqrt(90)/20 * np.sin(omega * t + phi)

plt.figure(figsize=(10, 5))
plt.plot(t, I_t,color='green', label='I(t) = √90/20sin(2t + -71.5°)')
plt.title('I(t)')
plt.xlabel('time(second)')
plt.ylabel('currunt(A)')
plt.grid()
plt.axhline(0, color='black', lw=0.5, ls='--')
plt.legend()
plt.show()
```
