## Level 2: Multiple Linear Equations
now let's look at this simple electronic circuit with this variable:

$V_1=10$

$I_2=\frac 1{20} v_1$

$I_3=3$

```python
import schemdraw
import schemdraw.elements as elm

with schemdraw.Drawing() as d:
    d.config(unit=5)
    V1 = elm.SourceV().label('10V')
    R1 = elm.Line().right()
    elm.Dot() 
    d.push()
    R2 = elm.Resistor().down().label('20Ω', loc='bot', rotate=True)
    elm.CurrentLabel(top=False, ofst=.3).at(R2).label('$I_2$')
    elm.Dot()
    d.pop()
    L1 = elm.Line()
    I1 = elm.SourceI().down().label('3A', loc='bot').reverse()
    L2 = elm.Line().tox(V1.start)
    elm.LoopCurrent([R1,R2,L2,V1], pad=1.25).label('$I_1$')
    elm.LoopCurrent([R1,I1,L2,R2], pad=1.25).label('$I_3$').reverse() 
```

## Level 3: Multiple Linear Equations with Changing Input 

continue our example from above, but an idea in this level is what happend if our input always change with time , it means what if our input change as a function of time ?

For this purpose let's suppose :

$ V_1(t)=10 * sin(\pi t)$

in this case our system are memoryles,which means nothing depends on any previous value of the state of the system.

and with this input our circuit change:




```python

import schemdraw
import schemdraw.elements as elm

with schemdraw.Drawing() as d:
    d.config(unit=5)
    V1 = elm.SourceSin().label('$V_1(t) = 10 +\\sin(\\pi t)$')
    R1 = elm.Line().right()
    elm.Dot() 
    d.push()
    R2 = elm.Resistor().down().label('20Ω', loc='bot', rotate=True)
    elm.CurrentLabel(top=False, ofst=.3).at(R2).label('$I_2$')
    elm.Dot()
    d.pop()
    L1 = elm.Line()
    I1 = elm.SourceI().down().label('3A', loc='bot').reverse()
    L2 = elm.Line().tox(V1.start)
    elm.LoopCurrent([R1,R2,L2,V1], pad=1.25).label('$I_1$')
    elm.LoopCurrent([R1,I1,L2,R2], pad=1.25).label('$I_3$').reverse() 
```

```python
from PySpice.Probe.Plot import plot
from PySpice.Spice.Library import SpiceLibrary
from PySpice.Spice.Netlist import Circuit
from PySpice.Unit import *


circuit = Circuit('Sine Voltage Source Example')


circuit.SinusoidalVoltageSource(1,'input' , circuit.gnd, amplitude=1@u_V, frequency=50@u_Hz, offset=10@u_V)


circuit.R(1, 'input', circuit.gnd, 1@u_kΩ)

circuit.I(1,  circuit.gnd,'input', 1@u_kΩ)

simulator = circuit.simulator(temperature=25, nominal_temperature=25)
analysis = simulator.transient(step_time=1@u_ms, end_time=50@u_ms)


import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plot(analysis['input'])
plt.title('Sinusoidal Voltage Source')
plt.xlabel('Time [s]')
plt.ylabel('Voltage [V]')
plt.grid()
plt.show()
```

as picture shows our inputs always change by time and in this way we should change our equation:

$I_1(t)=\frac 1{20}V_1(t) -3 $

and if we replace $V_1(t)$ by $10 * sin(\pi t)$ equation convert to this :

$I_1=\frac 12 sin(\pi t) - 3$

The big idea here is that if the system obeys certain properties, then we can broaden our thinking: instead of reasoning about our system as a function

$x \to F(x) \to y $ 

where x,y are numbers, to now thinking about our system as a capital-S System which takes a signal (a function of time) in and outputs another signal, 

$x(t) \to F(x(t)) \to y(t) $ 


## Level 4: Frequency Domain Addition 

now let's think what happend if more than one input signal effect on outputs , for example lets think $V_1$ is:

$V_1=10+sin(\pi t) +(0.1sin(10\pi t))$

and the circuit of this is:

```python

import schemdraw
import schemdraw.elements as elm

with schemdraw.Drawing() as d:
    d.config(unit=5)
    V1 = elm.SourceSin().label('$V_1(t) = 10 +\\sin(\\pi t) + (0.1\\sin(10\\pi t)) $')
    R1 = elm.Line().right()
    elm.Dot() 
    d.push()
    R2 = elm.Resistor().down().label('20Ω', loc='bot', rotate=True)
    elm.CurrentLabel(top=False, ofst=.3).at(R2).label('$I_2$')
    elm.Dot()
    d.pop()
    L1 = elm.Line()
    I1 = elm.SourceI().down().label('3A', loc='bot').reverse()
    L2 = elm.Line().tox(V1.start)
    elm.LoopCurrent([R1,R2,L2,V1], pad=1.25).label('$I_1$')
    elm.LoopCurrent([R1,I1,L2,R2], pad=1.25).label('$I_3$').reverse() 
```

```python
from PySpice.Probe.Plot import plot
from PySpice.Spice.Library import SpiceLibrary
from PySpice.Spice.Netlist import Circuit
from PySpice.Unit import *


circuit = Circuit('Sine Voltage Source Example')


circuit.V(1, 'input', 'node1', 10@u_V)


circuit.SinusoidalVoltageSource(2, 'node1', 'node2',
                                 amplitude=1@u_V,
                                 frequency=5@u_Hz)  


circuit.SinusoidalVoltageSource(3, 'node2', circuit.gnd,
                                 amplitude=0.1@u_V,
                                 frequency=50@u_Hz)  


circuit.R(1, 'input', circuit.gnd, 1@u_kΩ)

circuit.I(1,  circuit.gnd,'input', 1@u_kΩ)

simulator = circuit.simulator(temperature=25, nominal_temperature=25)
analysis = simulator.transient(step_time=5@u_ms, end_time=1000@u_ms)


import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plot(analysis['input'])
plt.title('Sinusoidal Voltage Source')
plt.xlabel('Time [s]')
plt.ylabel('Voltage [V]')
plt.grid()
plt.show()
```

The big idea here is that if we take a linear system and put in the sum of two sine waves at two different frequencies, our output will also have the sum of two sine ways at those same two frequencies.

## Level 5: A Single Differential Equation 

in level 3 we talk about memoryles system ,but in many real systems we need some differential equation like $\frac {dx}{dt}$ and cause in differential equation we need to now the value of last time so our systems should have memory.
Taking a derivative is a linear operator,this should be weird .
for example:

$\frac {dx}{dt} sin(t)=cos(t)$

Since $sin(t)$ and $cos(t)$ aren’t proportional to each other, how are they possibly linear?

based on definition of linearity in function :

if F is linear then:

$F(a.x(t) + b.y(t)) = a.F(x(t)) + b.F(y(t))$

A linear operator scales up and down in size with a constant term, and it distributes linearly to a sum.

A derivative follows these properties, so a derivative is a linear operator.

The big idea here is that when we talk about linear systems, we’re not talking about just mapping an input value (a single number) to an output value (a different number). We’re really talking about something that takes an input signal (a function of time) and gives an output signal (a different function of time). We’ve defined linearity more rigorously, and any linear system as we’ve previously defined in terms of $Gw=H$
 is a linear system regardless of how we define inputs or outputs within $w$

## Level 6: Multiple Differential Equations 

by combine level 2 and level 5 , we can take the general matrix equation form :

$Gw=H$

and if we extend it :

$Ax+B\dot x=H$

in this equation $\dot x$ is the matrix of derviation of each individual component :

$$
\dot x
=
\begin{bmatrix}
\frac {dx}{dt} x_1(t) \\
\frac {dx}{dt} x_2(t)\\
\vdots \\
\frac {dx}{dt} x_n(t)

\end{bmatrix}
$$

this is the most general form of matrix equation , and if we don't need to use all derivative component we can set amount of B to 0.


To solve the equation $Gw = H $ uniquely, we need a square $ N \times N $ matrix $ G $, meaning the number of unknowns must equal the number of linearly independent variables. However, when differential equations are involved, there’s a complication: both $ x_i $ and its derivative $ \dot{x}_i $ are unknowns. 

If the original matrices $ A $ and $ B $ are each $ N \times N $, the combined matrix $ G = [A \, B] $ becomes $ N $ rows by $ 2N $ columns, leading to more unknowns than equations. This means the system is no longer square, reflecting the need for additional information (e.g., $ x $ to compute $\dot{x}$ or vice versa).

To solve a Kth-order linear differential equation, K additional constraints (e.g., initial or boundary values) are needed. These constraints select a unique solution from the infinitely many possibilities allowed by the differential equation.

The big idea here is that we can easily modify our approach to systems of linear equations to incorporate linear differential equations, which occur in electronics every time there’s a capacitor or inductor.
